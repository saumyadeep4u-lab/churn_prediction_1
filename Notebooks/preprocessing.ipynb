## Data Preprocessing

- Data Type Correction
- Duplicate Handling
- Remove irrelevant columns
- Save cleaned data

```python
# Imports
import pandas as pd
import numpy as np
```

```python
# Load data
DATA_PATH = r"C:\Users\hp\Downloads\BIA\Customer_Churn _Prediction\Data\Telco_Customer_Churn.csv"
data = pd.read_csv(DATA_PATH)
data.head()
```

```python
#DATA backup
df= data.copy()
df.info()
```

```python
# Meta Data
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
all_columns = df.columns.to_list()
num_columns = df.select_dtypes(include = "number").columns.to_list()
cat_columns = df.select_dtypes(include= "object").columns.to_list()
print("\nAll columns: \n",all_columns)
print("\nNumeric columns: \n", num_columns)
print("\nCategorical Columns: \n",cat_columns)
print("\nData type: \n",df.dtypes)
```

```python
# Duplicate
print("Duplicate details:", df.duplicated().sum())
print("Duplicate rows: \n")
df[df.duplicated()]
```

```python
# Drop irrelevant column
df.drop(columns = ["customerID"], axis = 1, inplace= True)
```

```python
# Change data type of "TotalCharges"
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors = "coerce")
df["TotalCharges"].dtype
```

```python
# Encode "Churn" Column
df["Churn_encoded"]= df["Churn"].map({"Yes":1, "No":0})
df.head()
```

```python
# Save cleaned data
CLEANED_DATA_PATH = r"C:\Users\hp\Downloads\BIA\Customer_Churn _Prediction\Data\cleaned_data.csv"
df.to_csv(CLEANED_DATA_PATH, index = False)
```
