# Exploratory Data Analysis
- Statistical Analysis
- Visualization

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
CLEANED_DATA_PATH = r"C:\Users\hp\Downloads\BIA\Customer_Churn _Prediction\Data\cleaned_data.csv"
data = pd.read_csv(CLEANED_DATA_PATH)
data.head()
```

```python
# Statistical Analysis
df= data.copy()
df.describe()
```

```python
# Categorical & Numeric Columns
cat_cols = df.select_dtypes(include = "object").columns.to_list()
num_cols = df.select_dtypes(include = "number").columns.to_list()
cat_cols.remove("Churn")
print("Categorical Columns: \n", cat_cols)
print("\nNumeric Columns: \n", num_cols)
```

```python
# Target distribution
plt.figure(figsize = (10,6))
ax = sns.countplot(x = "Churn", data = df)
# Bar label
for container in ax.containers:
    ax.bar_label(container)
```

```python
# Numeric column Vs. Churn
for col in num_cols:
    plt.figure(figsize = (10,6))
    sns.violinplot(y= col, hue = "Churn", data= df)
    plt.xlabel(f"Churn")
    plt.ylabel(f"Distribution of {col}")
    plt.title(f"Churn vs {col}")
    plt.show()
```

```python
# Churn Vs. Categorical Columns
for col in cat_cols:
    plt.figure(figsize = (10,6))
    ax = sns.countplot(x = col, hue= "Churn", data = df)
    # bar label
    for container in ax.containers:
        ax.bar_label(container)
plt.xlabel(f"{col}")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.title(f"{col} Vs. Churn")
```

```python
# Corelation 
corr = df.corr(numeric_only = True)
print("The corelation map: \n")
sns.heatmap(corr, annot = True)
plt.show()
```
